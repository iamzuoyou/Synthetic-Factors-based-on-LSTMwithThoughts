```python
import numpy as np
import pandas as pd
```

```python
features = pd.read_csv("上证股票技术因子数据200101-241217.csv")
features.drop(features.columns[0],axis=1,inplace=True)
features.shape
```

```python
qfq_features = []
for f in features.columns:
    if(f == "ts_code" or f == "trade_date" or f == "pre_close" or f == "change" or f == "vol" or f == "amount" or f == "pct_chg"
       or f == "turnover_rate" or f == "turnover_rate_f" or f == "volume_ratio" or f == "pe" or f == "pe_ttm" or f == "pb" or f == "ps"
       or f == "ps_ttm" or f == "dv_ratio" or f == "dv_ttm" or f == "total_share" or f == "float_share" or f == "free_share"
       or f == "total_mv" or f == "circ_mv" or f == "adj_factor" or f == "downdays" or f == "updays" or f =="lowdays" or f == "topdays"):
        qfq_features.append(f)
    elif "qfq" in f:
        qfq_features.append(f)
len(qfq_features)
```

```python
features = features[qfq_features]
features.head()
```

```python
trade_days = features.groupby("ts_code").count().iloc[:,0].sort_values(ascending=False)
unfull_codelist = list(trade_days[trade_days < 1202].index)
len(unfull_codelist)
```

```python
for i in unfull_codelist:
    features.drop(features[features["ts_code"] == i].index,axis=0,inplace=True)
features.reset_index(drop=True,inplace=True)
features.shape
```

```python
nalist = features.isnull().sum().sort_values(ascending=False)
nalist
```

```python
features[nalist.index[1:3]] = features[nalist.index[1:3]].fillna(0) #pe NaN values are filled by 0, due to negative values were recorded as NaN
features.drop(nalist.index[0],axis=1,inplace=True) #drop columns with too many NaN values
features.drop(nalist.index[3],axis=1,inplace=True)
features.head()
```
